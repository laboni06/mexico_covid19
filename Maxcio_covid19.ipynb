### Remove all columns between a specific column to another

#### Drop columns from a Dataframe using iloc[] and drop() method

```python
df
```

```python
drop_colmn = df.drop(df.loc[:, 'ENTITY_RES':'ENTITY'].columns ,axis = 1)
drop_colmn
```

#### [NOTE: Difference between loc() & iloc() is iloc() exclude last column range element ]
